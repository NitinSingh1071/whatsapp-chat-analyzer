```python
import re
import pandas as pd
```

```python
f = open('BDA.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = "\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s"
```

```python
date = re.findall(pattern, data)
date
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
df = pd.DataFrame({'user_messages':messages, 'messages_date':date})
```

```python
df.head()
```

```python
df['messages_date'] = pd.to_datetime(df['messages_date'], format='%m/%d/%y, %H:%M - ')
df.rename(columns={'messages_date':'date'}, inplace=True)
```

```python
df.shape
```

```python
user = []
messages = []

for message in df['user_messages']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: #usernames
        user.append(entry[1])
        messages.append(entry[2])
    else:
        user.append("group_notification")
        messages.append(entry[0])
```

```python
df['user'] = user
df['messages'] = messages
df.drop(columns=['user_messages'], inplace=True)
```

```python
df.sample(20)
```

```python
from collections import Counter
```

```python
temp = df[df['user'] != 'group_notification']
temp = temp[temp['messages'] != '<Media omitted>\n']
f = open('stop_hinglish.txt', 'r')
stop_word = f.read()

words = []
for message in temp['messages']:
    for word in message.lower().split():
        if word not in stop_word:
            words.append(word)

most_common_words_df = pd.DataFrame(Counter(words).most_common(20))
```
